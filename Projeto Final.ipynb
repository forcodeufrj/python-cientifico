Um trocador de calor é formado por um tubo, por onde é passado o fluido, envolvo por um casco, por onde é passado vapor saturado. O casco é alimentado pelo vapor, a fim de aquecer o fluido que escoa pelo tubo.

Para se obter o comprimento do trocador de calor, é necessário realizar a integração do balanço de energia do sistema, cuja equação resultante é a seguinte:

```python
from IPython.display import display, Math, Latex
display(Math(r'L = \frac{W}{\pi D}\int_{T_{entrada}}^{T_{saida}}\left [ \frac{C_P{(T)}}{h(T)(T_{vapor}-T)} \right ]dT'))
```

Sendo:
#### L: comprimento do trocador;
#### W: vazão mássica do fluido do tubo;
#### D: diâmetro do tubo;
#### T: temperatura;
#### Cp: calor específico do fluido do tubo;
#### h: coeficiente de transferência de calor entre o tubo e o casco.


O coeficiente h é obtido a partir da correlação empírica:


```python
display(Math(r'h(T)=\frac{0,023k(T)}{D}\left ( \frac{4W}{\pi D\mu (T)} \right )^{0,8}\left ( \frac{\mu (T)C_P{(T)}}{k(T)} \right )^{0,4}'ki))
```

Sendo:
#### k: coeficiente de condutividade térmica do fluido do casco;
#### µ: viscosidade de condutividade térmica do fluido do casco.
